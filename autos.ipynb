```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("autos.csv",encoding="latin")
```

```python
df.columns
```

```python
def clean_columns(col):
    col=col.strip()
    col=col.replace("(","")
    col=col.replace(")","")
    col=col.replace(" ","_")
    return col
```

```python
new_columns=[]
for name in  df.columns:
    x=clean_columns(name)
    new_columns.append(x)
```

```python
df.columns=new_columns
```

```python
df.columns
```

```python
df.head(3)
```

```python
df["postalCode"].unique()
```

```python
# Convert the column to string type first to ensure .str accessor works
df["powerPS"] = df["powerPS"].astype(str).str.replace('"', "").astype("float")
```

```python
df.info()
```

```python
df["price"]=df["price"].str.replace("INR","").str.replace("$","").str.replace(",","").astype("int")
```

```python
df.info()
```

```python
df.rename(columns={"nrOfPictures":"NumOfPictures",\
                   "monthOfRegistration": "MoR"})
```
